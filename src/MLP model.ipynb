```python
frequency_1daybefore = dataset['Frequency'].shift(+1)
frequency_2daybefore = dataset['Frequency'].shift(+2)
frequency_3daybefore = dataset['Frequency'].shift(+3)
#frequency_4daybefore = dataset['frequency'].shift(+4)
#frequency_5daybefore = dataset['frequency'].shift(+5)
#frequency_6daybefore = dataset['frequency'].shift(+6)
#frequency_7daybefore = dataset['frequency'].shift(+7)
```

```python
df['frequency_1daybefore']  = frequency_1daybefore.tolist()
df['frequency_2daybefore']  = frequency_2daybefore.tolist()
df['frequency_3daybefore']  = frequency_3daybefore.tolist()
#df['frequency_4daybefore']  = frequency_4daybefore.tolist()
#df['frequency_5daybefore']  = frequency_5daybefore.tolist()
#df['frequency_6daybefore']  = frequency_6daybefore.tolist()
#df['frequency_7daybefore']  = frequency_7daybefore.tolist()
df
```

```python
df = df.fillna(df.mean())
df.info()
```

```python
df
```

```python
X_train = df.loc[df.index < '2011-08-01']
X_train = X_train.drop('Frequency', axis = 1)
X_train
```

```python
Y_train = df.loc[df.index < '2011-08-01']
Y_train = Y_train.drop(labels = ['frequency_1daybefore','frequency_2daybefore','frequency_3daybefore'] , axis = 1)
Y_train
```

```python
X_test = df.loc[df.index >= '2011-08-01']
X_test = X_test.drop('Frequency', axis = 1)
X_test
```

```python
Y_test = df.loc[df.index >= '2011-08-01']
Y_test = Y_test.drop(labels = ['frequency_1daybefore','frequency_2daybefore','frequency_3daybefore'] , axis = 1)
Y_test1 = Y_test
Y_test
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
```

```python
model.add(Dense(100, activation ='relu', input_dim = 3 ))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X_train, Y_train, epochs= 2000 ,verbose= 0)
```

```python
prediction = model.predict(X_test, verbose = 0)
prediction
```

```python
prediction = pd.DataFrame(prediction, columns=['frequency1'])
prediction 
```

```python
prediction = prediction['frequency1']
```

```python
Y_test1['frequency1']  = prediction.tolist()
Y_test1 = Y_test1.drop('Frequency', axis = 1)
Y_test1
```

```python
Y_test = Y_test.drop('frequency1', axis = 1)
Y_test
```

```python
plt.rcParams["figure.figsize"] = (14,7)
plt.plot(Y_test, 'g' ,label= 'Actual Trend')
plt.plot(Y_test1, 'r--' ,label= 'Prediction Trend')
plt.grid()
plt.xlabel('Next 30 days', fontsize= 12, weight='bold')
plt.ylabel('Number of frequncy of Patients', fontsize= 12, weight='bold')
plt.title('Results of prediction using MLP', fontsize= 14, weight='bold')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import sklearn.metrics as metrics
import numpy as np
```

```python
m = mean_squared_error(Y_test,Y_test1)
print(f"MSE : {m:.5f}")
rmse = np.sqrt(m)
print(f"RMSE : {rmse:.5f}")
mae = metrics.mean_absolute_error(Y_test,Y_test1)
print(f"MAE : {mae:.5f}")
r2 = r2_score(Y_test,Y_test1)
print(f"r2 : {r2:.5f}")
```
