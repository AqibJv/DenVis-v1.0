```python
#STL

import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import register_matplotlib_converters
from datetime import datetime
```

```python
register_matplotlib_converters()
Total = df['Age'].resample('D').count() #day of week effect
Total
```

```python
from statsmodels.tsa.seasonal import STL
result = STL(Total).fit()
```

```python
seasonal, trend, resid= result.seasonal, result.trend, result.resid
# Calculate the day-of-week component
Dt = Total - seasonal
Dt_weekly = Dt.resample('W').mean()
Dt_dayofweek = Dt.groupby(Dt.index.dayofweek).mean()
```

```python
plt.figure(figsize = (20,10))

plt.subplot(5,1,1)
plt.plot(Total)
plt.title('original series', fontsize = 16)
plt.xlabel("Date")
plt.ylabel("No. of Patients")

plt.subplot(5,1,2)
plt.plot(trend)
plt.title('Trend, Tt', fontsize = 16)
plt.xlabel("Date")
plt.ylabel("No. of Patients")

plt.subplot(5,1,3)
plt.plot(seasonal)
plt.title('Seasonal, St', fontsize = 16)
plt.xlabel("Date")
plt.ylabel("No. of Patients")

plt.subplot(5,1,4)
plt.plot(Dt_dayofweek)
plt.title('Day of Week, Dt', fontsize = 16)
plt.xlabel("Date")
plt.ylabel("No. of Patients")

plt.subplot(5,1,5)
plt.plot(resid)
plt.title('Noise, Nt', fontsize = 16)
plt.xlabel("Date")
plt.ylabel("No. of Patients")

plt.tight_layout()

```

```python
estimated = trend + seasonal
plt.figure(figsize=(12,4))
plt.plot(Total, label = 'Actual Time Series')
plt.plot(estimated, label = 'Estimated = Trend + Seasonal')
plt.legend(frameon=False)
plt.xlabel("Date")
plt.ylabel("No. of Patients")

#blue is original while orange is estimated
```

## Anomaly Detection

```python
resid_mu = resid.mean()
resid_dev = resid.std()

lower = resid_mu - 2*resid_dev
upper = resid_mu + 2*resid_dev
```

```python
upper
```

```python
plt.figure(figsize=(10,4))
plt.plot(resid)
plt.xlabel("Date")
plt.ylabel("No. of Patients")
plt.title('Upper and Lower limits for normal time series')

plt.fill_between([datetime(2011,5,1), datetime(2022,12,19)], lower, upper, color='g', alpha=0.25, linestyle='--', linewidth=2)
plt.xlim(datetime(2011,4,15), datetime(2011,9,15))
```

```python
anomalies = Total[(resid < lower) | (resid > upper)]
```

```python
plt.figure(figsize=(10,4))
plt.plot(Total)
plt.xlabel("Date")
plt.ylabel("No. of Patients")
    
plt.scatter(anomalies.index, anomalies, color='r', marker='D')
plt.title('Anomalies in red diamond')
plt.xticks(rotation=-10) # Rotate x-axis labels by 45 degrees
plt.show() # Display the plot
```

### Checking stationarity 

```python
from statsmodels.tsa.stattools import adfuller
# Run Augmented Dickey Fuller Test
result = adfuller(Total)
# Print p-value
print(result[1])
if result[1]<0.5:
    print('Time Series is stationary')
    
else:
    print('There is a need to make this Time Series stationary')
```
